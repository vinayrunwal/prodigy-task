```python
import pandas as pd

# Now this should work
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the year and some countries (you can modify these)
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil']
year = '2020'

# Filter the data for selected countries and the chosen year
data_2020 = df[df['Country Name'].isin(selected_countries)][['Country Name', year]]

# Rename columns for clarity
data_2020.columns = ['Country', 'Population']

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=data_2020, x='Country', y='Population', palette='viridis')

plt.title(f'Total Population in {year} for Selected Countries')
plt.ylabel('Population')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
